## Ejercicio 2: Creaci√≥n de un ciclo "for"

```python
new_planet = ''
planets = []

while(new_planet.lower() != 'done'):
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Ingrese un planeta o \'done\' para salir')


for i in planets:
    print(i)
```
