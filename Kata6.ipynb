## Ejercicio 2: Trabajando con datos de una lista

```python
Planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
Planeta = input('Ingrese el nombre del planeta iniciando con mayúscula')

planet_index = Planetas.index(Planeta)



print('Estos planetas estan más cerca que ' + Planeta)
print(Planetas[0:planet_index])



print('Estos planetas estan más lejos que ' + Planeta)
print(Planetas[planet_index + 1:])
```
