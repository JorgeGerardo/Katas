## Ejercicio 1: Trabajar con argumentos en funciones

```python
def GenerarReporte(TanquePrincipal, TanqueExterno, TanqueHydrogeno):
    total_average = (TanquePrincipal + TanqueExterno + TanqueHydrogeno) / 3
    return f"""Fuel Report:
    Promedio total: {total_average}%
    Tanque principal: {TanquePrincipal}%
    Tanque externo: {TanqueExterno}%
    Tanque de hidrogeno: {TanqueHydrogeno}% 
    """

print(GenerarReporte(80, 70, 85))


def average(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items
    

average([80, 85, 81]) 

def GenerarReporte(TanquePrincipal, TanqueExterno, TanqueHydrogeno):
    return f"""Fuel Report:
    Total Average: {average([TanquePrincipal, TanqueExterno, TanqueHydrogeno])}%
    Tanque principal: {TanquePrincipal}%
    Tanque externo: {TanqueExterno}%
    Tanque de hidrogeno: {TanqueHydrogeno}% 
    """

print(GenerarReporte(88, 76, 70))
```

## Ejercicio 2: Trabajo con argumentos de palabra clave

```python
def mission_report(destination, *minutes, **fuel_reservoirs):
    return f"""
    Mision a {destination}
    Tiempo total del viaje: {sum(minutes)} min
    Cuanto combustible queda: {sum(fuel_reservoirs.values())}
    """

print(mission_report("Luna", 10, 15, 51, main=300000, external=200000))


def mission_report(destination, *minutes, **fuel_reservoirs):
    main_report = f"""
    Misión a {destination}
    Tiempo total del viaje: {sum(minutes)} min
    Combustible restante: {sum(fuel_reservoirs.values())}
    """
    for tank_name, gallons in fuel_reservoirs.items():
        main_report += f"{tank_name} tanque: {gallons} galones restantes\n"
    return main_report

print(mission_report("Luna", 8, 11, 55, main=300000, external=200000))
```
